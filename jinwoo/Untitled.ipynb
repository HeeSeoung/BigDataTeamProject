```python
import pandas as pd
```

```python
nation = pd.read_csv('./data/olympic_results.csv')
rio = pd.read_csv('./data/rio_athletes.csv', index_col=0)
```

```python
nation
```

```python
codes = list(nation['country_3_letter_code'].unique())
code_dict = dict()

for code in codes:
    mask = nation['country_3_letter_code'] == code
    temp = nation[mask]
    nation_name = temp.iloc[0, 8]
    code_dict[code] = nation_name
code_dict['ESP']
```

```python
def mapping(x):
    try:
        return code_dict[x]
    except:
        return None
```

```python
rio['Nation'] = rio['nationality'].apply(mapping)
rio
```

```python
SIN_mask = rio['nationality'] == 'SIN'
rio.loc[SIN_mask, 'Nation'] = 'Singapore'

LIB_mask = rio['nationality'] == 'LIB'
rio.loc[LIB_mask, 'Nation'] = 'Lebanon'
```

```python
rio.to_csv('./data/rio_athletes.csv')
```
