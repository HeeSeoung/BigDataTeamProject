```python
data = pd.read_csv('./data/olympic.csv')
data
```

```python
summer = data[data['type'] == 'summer']
winter = data[data['type'] == 'winter']
```

```python
mask = summer['Nation'] == 'Japan'
japan = summer[mask]
japan = japan.set_index('year')
japan = japan.sort_values('year')
japan
```

```python
sns.lineplot(data=japan, x='year', y='rank', marker='o')
```

```python
sns.lineplot(data=japan, x='year', y='total_rank', marker='o')
```

```python
sns.lineplot(data=japan, x='year', y='Gold', marker='o')
```

```python
japan[['Gold', 'total', 'total_rank', 'rank']].diff()
```

```python
sns.lineplot(data=UK, x='year', y='total_rank', marker='o')
```
