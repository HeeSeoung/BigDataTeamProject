```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./data/olympic.csv')
data
```

```python
summer_mask = data['type'] == 'summer'
summer = data[summer_mask]

japan_mask = summer['Nation'] == 'Japan'
UK_mask = summer['Nation'] == 'Great Britain'

japan = summer[japan_mask]
UK = summer[UK_mask]

japan = japan.set_index('year')
UK = UK.set_index('year')
```

```python
japan
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x=japan.index, y=japan['Gold'], marker='o', label='Japan')
sns.lineplot(x=UK.index, y=UK['Gold'], marker='o', label='UK')
plt.axvline(x=2012, color='r', linestyle='--', linewidth=1)
plt.axvline(x=2020, color='b', linestyle='--', linewidth=1)
plt.xticks(list(range(1996, 2021, 4)))
plt.legend()
plt.grid()
plt.title('UK VS JAPAN / GLOD medals')
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x=japan.index, y=japan['rank'], marker='o', label='Japan')
sns.lineplot(x=UK.index, y=UK['rank'], marker='o', label='UK')
plt.axvline(x=2012, color='r', linestyle='--', linewidth=1)
plt.axvline(x=2020, color='b', linestyle='--', linewidth=1)
plt.xticks(list(range(1996, 2021, 4)))
plt.legend()
plt.grid()
plt.title('UK VS JAPAN / Ranking')
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x=japan.index, y=japan['total_rank'], marker='o', label='Japan')
sns.lineplot(x=UK.index, y=UK['total_rank'], marker='o', label='UK')
plt.axvline(x=2012, color='r', linestyle='--', linewidth=1)
plt.axvline(x=2020, color='b', linestyle='--', linewidth=1)
plt.xticks(list(range(1996, 2021, 4)))
plt.legend()
plt.grid()
plt.title('UK VS JAPAN / Total Ranking')
```

```python
USA_mask = summer['Nation'] == 'United States of America'
USA = summer[USA_mask]
USA = USA.set_index('year')

china_mask = summer['Nation'] == 'People\'s Republic of China'
china = summer[china_mask]
china = china.set_index('year')
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x=USA.index, y=USA['Gold'], marker='o', label='USA')
sns.lineplot(x=china.index, y=china['Gold'], marker='o', label='China')
plt.xticks(list(range(1996, 2021, 4)))
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x=USA.index, y=USA['rank'], marker='o', label='USA')
sns.lineplot(x=china.index, y=china['rank'], marker='o', label='China')
plt.xticks(list(range(1996, 2021, 4)))
plt.legend()
plt.yticks(list(range(1, 5)))
plt.grid()
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x=USA.index, y=USA['total_rank'], marker='o', label='USA')
sns.lineplot(x=china.index, y=china['total_rank'], marker='o', label='China')
plt.xticks(list(range(1996, 2021, 4)))
plt.legend()
plt.yticks(list(range(1, 5)))
plt.grid()
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x=japan.index, y=japan['Gold'], marker='o', label='Japan')
sns.lineplot(x=china.index, y=china['Gold'], marker='o', label='China')
sns.lineplot(x=UK.index, y=UK['Gold'], marker='o', label='UK')
plt.axvline(x=2008, color='orange', linestyle='--', linewidth=1)
plt.axvline(x=2020, color='b', linestyle='--', linewidth=1)
plt.axvline(x=2012, color='g', linestyle='--', linewidth=1)
plt.xticks(list(range(1996, 2021, 4)))
plt.legend()
plt.grid()
plt.title('GLOD medals Home Advantage', size=15)
```
